```python
import tensorflow as tf
from tensorflow.keras import layers, models
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt

# ===============================
# Cargar dataset desde TFDS
# ===============================
ds, info = tfds.load("oxford_iiit_pet:4.0.0", with_info=True)

print("Clases:", info.features["label"].names)
print("Ejemplos train (completo):", info.splits["train"].num_examples)

# ===============================
# Usar solo 1000 ejemplos
# ===============================
N = 1000
train_size = int(0.8 * N)

raw_ds = ds["train"].take(N)

# ===============================
# Preprocesamiento
# ===============================
IMG_SIZE = 64
NUM_CLASSES = 3   # fondo, borde, animal

def normalize_img_mask(sample):
    img = tf.image.resize(sample["image"], (IMG_SIZE, IMG_SIZE)) / 255.0
    mask = tf.image.resize(sample["segmentation_mask"], (IMG_SIZE, IMG_SIZE), method="nearest")
    mask = tf.cast(mask, tf.int32) - 1   # {1,2,3} -> {0,1,2}
    return img, tf.squeeze(mask, -1)

raw_ds = raw_ds.map(normalize_img_mask, num_parallel_calls=tf.data.AUTOTUNE)

train_ds = raw_ds.take(train_size).shuffle(500).batch(32).repeat().prefetch(tf.data.AUTOTUNE)
val_ds   = raw_ds.skip(train_size).batch(32).prefetch(tf.data.AUTOTUNE)

steps_per_epoch = train_size // 32
val_steps = (N - train_size) // 32

# ===============================
# Mini U-Net
# ===============================
def conv_block(x,f):
    x = layers.Conv2D(f,3,padding="same",activation="relu")(x)
    x = layers.Conv2D(f,3,padding="same",activation="relu")(x)
    return x

def encoder_block(x,f):
    c = conv_block(x,f)
    p = layers.MaxPooling2D((2,2))(c)
    return c,p

def decoder_block(x,skip,f):
    us = layers.Conv2DTranspose(f,2,strides=2,padding="same")(x)
    return conv_block(layers.Concatenate()([us,skip]),f)

def build_unet(input_shape=(IMG_SIZE,IMG_SIZE,3), num_classes=NUM_CLASSES):
    inputs = layers.Input(input_shape)
    c1,p1 = encoder_block(inputs,32)
    c2,p2 = encoder_block(p1,64)
    c3,p3 = encoder_block(p2,128)
    bn = conv_block(p3,256)
    d1 = decoder_block(bn,c3,128)
    d2 = decoder_block(d1,c2,64)
    d3 = decoder_block(d2,c1,32)
    outputs = layers.Conv2D(num_classes,1,activation="softmax")(d3)
    return models.Model(inputs,outputs)

model = build_unet()
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()

# ===============================
# Entrenamiento
# ===============================
EPOCHS = 5 #Aumentar para obtener máscaras mas precisas.
history = model.fit(train_ds, epochs=EPOCHS, steps_per_epoch=steps_per_epoch,
                    validation_data=val_ds, validation_steps=val_steps)

# ===============================
# Visualizar predicciones
# ===============================
for imgs, masks in val_ds.take(1):
    sample_img = imgs[0].numpy()
    sample_mask = masks[0].numpy()
    pred_mask = tf.argmax(model.predict(sample_img[None,...])[0], axis=-1)

    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1); plt.imshow(sample_img); plt.title("Imagen")
    plt.subplot(1,3,2); plt.imshow(sample_mask); plt.title("Máscara Real")
    plt.subplot(1,3,3); plt.imshow(pred_mask); plt.title("Máscara Predicha")
    plt.show()
```
